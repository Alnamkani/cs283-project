```python
predictions_file_path = "/Users/alnamkani/Desktop/Not_Synced/CS/cs283-project/predictions.txt"

results = {}

with open(predictions_file_path, 'r') as file:
    for line in file:
        line = line.strip()
        if not line:
            continue
        parts = line.split('\t')
        results[parts[0]] = parts[1]
```

```python
len(results)
```

```python
import matplotlib.pyplot as plt

# Convert values to floats
accuracies = [float(val) for val in results.values()]

plt.figure(figsize=(12,6))
plt.hist(accuracies, bins=50, edgecolor='black')
plt.title('Distribution of Prediction Accuracies')
plt.xlabel('Accuracy')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Print some basic statistics
print(f"Average accuracy: {sum(accuracies)/len(accuracies):.4f}")
print(f"Min accuracy: {min(accuracies):.4f}")
print(f"Max accuracy: {max(accuracies):.4f}")
```

```python
# Count values above 0.9
above_90_percent = sum(1 for acc in accuracies if acc > 0.9)
print(f"Number of predictions above 0.9: {above_90_percent}")
print(f"Percentage above 0.9: {(above_90_percent/len(accuracies))*100:.1f}%")
```
